```python
import os, re, pandas as pd

path = os.path.expanduser("~/Desktop/community resiliance/useful.csv")
df = pd.read_csv(path)

df["tweet_clean"] = df["tweet"].astype(str)

def clean_text(s):
    s = s.lower()
    s = re.sub(r"http\S+|www\.\S+", " ", s)
    s = re.sub(r"@\w+", " ", s)
    s = s.replace("#", " ")
    s = re.sub(r"\s+", " ", s).strip()
    return s

df["tweet_clean"] = df["tweet_clean"].apply(clean_text)

# Save cleaned file
out = os.path.expanduser("~/Desktop/community resiliance/useful_clean.csv")
df.to_csv(out, index=False)

print("Cleaned file saved:", out)
print(df[["tweet","tweet_clean"]].head(10))
```

```python
import numpy as np
import pandas as pd

# Parse timestamps safely
df["_created_at"] = pd.to_datetime(df["_created_at"], errors="coerce")
df["_started_at"] = pd.to_datetime(df["_started_at"], errors="coerce")

# Use created_at as primary event time
event_time = df["_created_at"]

# Compute hours since event ONLY where time exists
max_time = event_time.max()
df["hours_since_event"] = (max_time - event_time).dt.total_seconds() / 3600

# Initialize recency_score as NaN
df["recency_score"] = np.nan

# Apply exponential decay ONLY to rows with valid timestamps
mask_valid_time = df["hours_since_event"].notna()
df.loc[mask_valid_time, "recency_score"] = np.exp(
    -0.1 * df.loc[mask_valid_time, "hours_since_event"]
)

# Clean up intermediate column
df.drop(columns=["hours_since_event"], inplace=True, errors="ignore")

# Sanity check
df["recency_score"].describe()
```

```python
import pandas as pd

# Make sure df exists first
df_model = df.copy()

print("df_model created. Shape:", df_model.shape)
```

```python
import numpy as np
import pandas as pd

def simplify_message_type(x):
    if pd.isna(x):
        return np.nan
    x = x.lower()
    if "not informative: personal" in x:
        return "personal"
    elif "informative" in x:
        return "informative"
    else:
        return "other"

# Create simplified label column
df_model["type_simple"] = df_model["type_of_message"].apply(simplify_message_type)

print(df_model["type_simple"].value_counts(dropna=False))
```

```python
train_df = df_model[
    df_model["type_simple"].notna() &
    (df_model["tweet_clean"].str.strip() != "")
].copy()

X_train = train_df["tweet_clean"].tolist()
y_train = train_df["type_simple"].tolist()

print("Training samples:", len(train_df))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

type_imputer = Pipeline([
    ("tfidf", TfidfVectorizer(
        max_features=30000,
        ngram_range=(1, 2),
        stop_words="english",
        min_df=5
    )),
    ("clf", LogisticRegression(
        max_iter=2000,
        class_weight="balanced",
        solver="lbfgs"
    ))
])

type_imputer.fit(X_train, y_train)
print("Model trained to learn message-type patterns")
```

```python
mask_missing = (
    df_model["type_simple"].isna() &
    (df_model["tweet_clean"].str.strip() != "")
)

X_missing = df_model.loc[mask_missing, "tweet_clean"].tolist()

print("Rows to impute:", len(X_missing))
```

```python
probs = type_imputer.predict_proba(X_missing)
preds = type_imputer.classes_[probs.argmax(axis=1)]
conf = probs.max(axis=1)
```

```python
CONF_THRESHOLD = 0.70  # safe choice

df_model.loc[mask_missing, "type_simple"] = np.where(
    conf >= CONF_THRESHOLD,
    preds,
    "other"
)

print(df_model["type_simple"].value_counts())
```

```python
# Create a final message type column
df_model["type_of_message_imputed"] = df_model["type_simple"]
```

```python
# Import transformer pipeline for LLM-based tweet classification
# and pandas/numpy for handling and storing tweet data and predictions
from transformers import pipeline
import pandas as pd
import numpy as np
```

```python
# Load a large pre-trained transformer (LLM) for zero-shot tweet classification
# Used to automatically label tweets as emergency, support offer, or neutral without manual annotation
zero_shot_classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)
```

```python
# Define target categories for automatic tweet classification:
# emergency requests, support offers, and neutral (general information)
candidate_labels = ["emergency", "support offer", "neutral"]
```

```python
# Automatically classify each tweet into emergency, support offer, or neutral
    # using a pre-trained LLM (BART) and return both the predicted label and confidence score
def zero_shot_label(text):
    if not isinstance(text, str) or text.strip() == "":
        return None, None

    result = zero_shot_classifier(
        text,
        candidate_labels,
        hypothesis_template="This tweet is about {}."
    )

    label = result["labels"][0] # Highest-confidence category
    score = result["scores"][0] # Confidence of the prediction

    return label, score
```

```python
# Automatically label each tweet as emergency, support offer, or neutral
# using a zero-shot LLM, and store both the predicted label and confidence score
df_model["auto_label"], df_model["auto_label_score"] = zip(*df_model["tweet_clean"].apply(zero_shot_label))
```

```python
# Check how many tweets were automatically labeled as emergency, support offer, or neutral by the LLM
df_model["auto_label"].value_counts()
```

```python
def finalize_label(row):
    label = row["auto_label"]
    score = row["auto_label_score"]

    if label == "emergency" and score >= 0.40:
        return "emergency"
    elif label == "support offer" and score >= 0.50:
        return "support offer"
    else:
        return "neutral"

df_model["auto_label_final"] = df_model.apply(finalize_label, axis=1)

df_model["auto_label_final"].value_counts()
```

```python
from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
import numpy as np
from sklearn.metrics import accuracy_score, f1_score

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=1)
    return {
        "accuracy": accuracy_score(labels, preds),
        "f1_macro": f1_score(labels, preds, average="macro")
    }
```

```python
# Task 1 dataset (emergency classification)

train_ds1 = Dataset.from_pandas(train_df[["tweet_clean", "label_task1"]])
val_ds1   = Dataset.from_pandas(val_df[["tweet_clean", "label_task1"]])
test_ds1  = Dataset.from_pandas(test_df[["tweet_clean", "label_task1"]])
```

```python
# Load RoBERTa tokenizer

roberta_tok = AutoTokenizer.from_pretrained("roberta-base")

def tokenize_task1(batch):
    return roberta_tok(
        batch["tweet_clean"],
        truncation=True,
        padding="max_length",
        max_length=128
    )

train_ds1 = train_ds1.map(tokenize_task1, batched=True)
val_ds1   = val_ds1.map(tokenize_task1, batched=True)
test_ds1  = test_ds1.map(tokenize_task1, batched=True)
```

```python
# Trainer expects column name "labels"

train_ds1 = train_ds1.rename_column("label_task1", "labels")
val_ds1   = val_ds1.rename_column("label_task1", "labels")
test_ds1  = test_ds1.rename_column("label_task1", "labels")

train_ds1.set_format("torch")
val_ds1.set_format("torch")
test_ds1.set_format("torch")
```

```python
# 3 classes for Task 1

model_task1 = AutoModelForSequenceClassification.from_pretrained(
    "roberta-base",
    num_labels=3
)
```

```python
training_args1 = TrainingArguments(
    output_dir="./roberta_task1",
    num_train_epochs=10,
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    weight_decay=0.01,
    report_to="none"
)
```

```python
trainer_task1 = Trainer(
    model=model_task1,
    args=training_args1,
    train_dataset=train_ds1,
    eval_dataset=val_ds1,
    compute_metrics=compute_metrics
)

trainer_task1.train()

```
